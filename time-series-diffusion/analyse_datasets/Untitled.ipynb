```python
freqs = list(np.linspace(0.05, 1, 20))
time = np.linspace(0, 2*np.pi, 1000)
data = [np.sin(2 * np.pi * freq * time) for freq in freqs]
```

```python
for i in range(20):
    plt.plot(time, data[i])
    plt.show()
    plt.close()
```

```python
data = np.array(data)
```

```python
data.shape
```

```python
intervals = [(0+i,50+i) for i in range(0,1000-49)]
intervals
```

```python
dataset = np.zeros((len(intervals)*20, 50))
for i in range(10):
    orig_dataset = data[i]
    for j in range(len(intervals)):
        min_ = intervals[j][0]
        max_ = intervals[j][1]
        
        idx = i*len(intervals)+j
        print(idx, min_, max_)
        dataset[idx] = orig_dataset[min_:max_]
```
